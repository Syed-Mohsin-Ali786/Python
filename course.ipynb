```python
import random
with open("words.txt", "r") as file:
    words = file.read().splitlines()

random_word = random.choice(words)
print(random_word)

blanks_arry = []
for blank in range(0, len(random_word)):
    blanks_arry.append("-")

guess_letter = input("Guess a letter :")

guess_word = ""


def replace_blanks():
    guess_word.join(blanks_arry)
    print(guess_word,"191")

def true(i):
    blanks_arry[i] = guess_letter
    replace_blanks()


def check():
    i = 0

    while guess_word == random_word:
        guess_letter = input("Guess a letter :")
        for letter in random_word:
            if guess_letter == letter:
                true(i)
            i += 1

        print(guess_word)


check()
```
