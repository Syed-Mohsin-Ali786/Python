```python
import matplotlib.pyplot as plt
import numpy as np

# Number of points for a 5-pointed star
# We can connect vertices of a regular pentagon in the order 0, 2, 4, 1, 3, 0
angles = np.array([0, 2, 4, 1, 3, 0]) * 2 * np.pi / 5 + np.pi / 2

xpoints = np.cos(angles)
ypoints = np.sin(angles)

plt.plot(xpoints, ypoints, marker='o')
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off') # remove axes for a clean look
plt.savefig('perfect_star.png', bbox_inches='tight')
plt.show()
plt.close()
```
